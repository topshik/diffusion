```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
file = open("time.txt", 'r')
data = file.read().split('\n')
steps = map(float, data[1][1:len(data[1]) - 1].split(', '))
time = map(float, data[2][1:len(data[2]) - 1].split(', '))
```

```python
plt.figure(figsize=(15, 6))
ax1 = plt.subplot(1, 2, 1)
ax2 = plt.subplot(1, 2, 2)
ax1.plot(steps, time, '.')
ax2.plot(steps[2:], time[2:])
plt.grid()
plt.show()
```
